```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('cars_engage_2022.csv')
```

```python
cars.head()
```

# Data Cleaning

```python
cars.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
cars.head()
```

```python
cars['Fuel_Type'].unique()
```

```python
cars['Make'].unique()
```

```python
cars['Model'].unique()
```

```python
cars['Ex-Showroom_Price'].unique()
```

```python
cars.info()
```

```python
cars['Ex-Showroom_Price'] = [i.replace('Rs. ','') for i in cars['Ex-Showroom_Price']]
cars['Ex-Showroom_Price'] = [i.replace(',','') for i in cars['Ex-Showroom_Price']]
cars['Ex-Showroom_Price'] = cars['Ex-Showroom_Price'].astype(int)
```

```python
cars.info()
```

```python
cars['Fuel_Tank_Capacity'] = cars['Fuel_Tank_Capacity'].astype(str)
cars['Fuel_Tank_Capacity'] = [i.replace('litres','') for i in cars['Fuel_Tank_Capacity']]
cars['Fuel_Tank_Capacity'] = cars['Fuel_Tank_Capacity'].astype(float)
```

```python
HorsePower = cars.Power.str.extract(r'(\d{1,4}).*').astype(int) * 0.98632
HorsePower = HorsePower.apply(lambda x: round(x,2))
cars.Power = HorsePower
```

```python
cars['City_Mileage'] = cars['City_Mileage'].astype(str)
cars['City_Mileage'] = [i.replace('km/litre','') for i in cars['City_Mileage']]
cars['City_Mileage'] = [i.replace('?','') for i in cars['City_Mileage']]
cars['City_Mileage'] = [i.replace(',','.') for i in cars['City_Mileage']]
cars['City_Mileage'] = [i.replace('12.5-12.7 ','12.6') for i in cars['City_Mileage']]
cars['City_Mileage'] = cars['City_Mileage'].astype(float)
cars['Highway_Mileage'] = cars['Highway_Mileage'].astype(str)
cars['Highway_Mileage'] = [i.replace('km/litre','') for i in cars['Highway_Mileage']]
cars['Highway_Mileage'] = [i.replace('8.3-8.4 ','8.4') for i in cars['Highway_Mileage']]
cars['Highway_Mileage'] = cars['Highway_Mileage'].astype(float)
```

```python
cars.info()
```

```python
cars['City_Mileage'] = cars['City_Mileage'].fillna(cars['City_Mileage'].mean())
cars['Highway_Mileage'] = cars['Highway_Mileage'].fillna(cars['Highway_Mileage'].mean())
cars['Fuel_Tank_Capacity'] = cars['Fuel_Tank_Capacity'].fillna(cars['Fuel_Tank_Capacity'].mean())
```

```python
cars.info()
```

```python
cars[cars['Make'].isna()]
```

```python
cars['Make'] = cars['Make'].fillna(cars["Model"].apply(lambda x: x.split()[0]))
```

```python
cars.info()
```

```python
cars['Make'].unique()
```

```python
cars
```

```python
cars.reset_index(drop = True)
```

```python
cars.describe()
```

```python
cars.to_csv("Cleaned_data.csv")
```

# Model

```python
X = cars.drop(columns = 'Ex-Showroom_Price')
y = cars['Ex-Showroom_Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['Make', 'Model', 'Variant', 'Fuel_Type']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['Make', 'Model', 'Variant', 'Fuel_Type']), remainder = 'passthrough')
```

```python
lm = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lm)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```
