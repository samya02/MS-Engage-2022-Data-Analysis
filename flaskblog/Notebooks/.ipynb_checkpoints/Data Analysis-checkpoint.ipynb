```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('cars_engage_2022 (3).csv')
```

```python
cars.head()
```

```python
cars.drop(columns= 'Unnamed: 0', inplace = True)
```

```python
cars['Ex-Showroom_Price'] = [i.replace('Rs. ','') for i in cars['Ex-Showroom_Price']]
cars['Ex-Showroom_Price'] = [i.replace(',','') for i in cars['Ex-Showroom_Price']]
cars['Ex-Showroom_Price'] = cars['Ex-Showroom_Price'].astype(int)
```

```python
cars['Displacement'] = cars['Displacement'].astype(str)
cars['Displacement'] = [i.replace(' cc','') for i in cars['Displacement']]
cars.loc[cars['Displacement'].isna(),'Displacement'] = np.nan
cars['Displacement'] = cars['Displacement'].astype(float)
```

```python
cars['Fuel_Tank_Capacity'] = cars['Fuel_Tank_Capacity'].astype(str)
cars['Fuel_Tank_Capacity'] = [i.replace('litres','') for i in cars['Fuel_Tank_Capacity']]
cars['Fuel_Tank_Capacity'] = cars['Fuel_Tank_Capacity'].astype(float)
```

```python
cars['Height'] = cars['Height'].astype(str)
cars['Height'] = [i.replace('mm','') for i in cars['Height']]
cars['Height'] = cars['Height'].astype(float)
cars['Length'] = cars['Length'].astype(str)
cars['Length'] = [i.replace('mm','') for i in cars['Length']]
cars['Length'] = cars['Length'].astype(float)
cars['Width'] = cars['Width'].astype(str)
cars['Width'] = [i.replace('mm','') for i in cars['Width']]
cars['Width'] = cars['Width'].astype(float)
```

```python
cars['City_Mileage'] = cars['City_Mileage'].astype(str)
cars['City_Mileage'] = [i.replace('km/litre','') for i in cars['City_Mileage']]
cars['City_Mileage'] = [i.replace('?','') for i in cars['City_Mileage']]
cars['City_Mileage'] = [i.replace(',','.') for i in cars['City_Mileage']]
cars['City_Mileage'] = [i.replace('12.5-12.7 ','12.6') for i in cars['City_Mileage']]
cars['City_Mileage'] = cars['City_Mileage'].astype(float)
cars['Highway_Mileage'] = cars['Highway_Mileage'].astype(str)
cars['Highway_Mileage'] = [i.replace('km/litre','') for i in cars['Highway_Mileage']]
cars['Highway_Mileage'] = [i.replace('8.3-8.4 ','8.4') for i in cars['Highway_Mileage']]
cars['Highway_Mileage'] = cars['Highway_Mileage'].astype(float)
```

```python
cars['ARAI_Certified_Mileage'] = cars['ARAI_Certified_Mileage'].astype(str)
cars['ARAI_Certified_Mileage'] = [i.replace('km/litre','') for i in cars['ARAI_Certified_Mileage']]
cars['ARAI_Certified_Mileage'] = [i.replace('22.4-21.9 ','22') for i in cars['ARAI_Certified_Mileage']]
cars['ARAI_Certified_Mileage'] = [i.replace('9.8-10.0 ','10') for i in cars['ARAI_Certified_Mileage']]
cars['ARAI_Certified_Mileage'] = [i.replace('kmpl ','') for i in cars['ARAI_Certified_Mileage']]
cars['ARAI_Certified_Mileage'] = cars['ARAI_Certified_Mileage'].astype(float)
cars['ARAI_Certified_Mileage_for_CNG'] = cars['ARAI_Certified_Mileage_for_CNG'].astype(str)
cars['ARAI_Certified_Mileage_for_CNG'] = [i.replace('km/kg','') for i in cars['ARAI_Certified_Mileage_for_CNG']]
cars['ARAI_Certified_Mileage_for_CNG'] = cars['ARAI_Certified_Mileage_for_CNG'].astype(float)
```

```python
cars['Kerb_Weight'] = cars['Kerb_Weight'].astype(str)
cars['Kerb_Weight'] = [i.replace('kg','') for i in cars['Kerb_Weight']]
cars['Kerb_Weight'] = [i.replace('1016-1043 ','1029') for i in cars['Kerb_Weight']]
cars['Kerb_Weight'] = [i.replace('1053-1080 ','1060') for i in cars['Kerb_Weight']]
cars['Kerb_Weight'] = cars['Kerb_Weight'].astype(float)
```

```python
cars['Ground_Clearance'] = cars['Ground_Clearance'].astype(str)
cars['Ground_Clearance'] = [i.replace(' mm','') for i in cars['Ground_Clearance']]
cars['Ground_Clearance'] = cars['Ground_Clearance'].astype(float)
HP = cars.Power.str.extract(r'(\d{1,4}).*').astype(int) * 0.98632
cars.Power =HP
```

```python
cars['Front_Track'] = cars['Front_Track'].astype(str)
cars['Front_Track'] = [i.replace(' mm','') for i in cars['Front_Track']]
cars['Front_Track'] = [i.replace(' (R13)','') for i in cars['Front_Track']]
cars['Front_Track'] = [i.replace(',','') for i in cars['Front_Track']]
cars['Front_Track'] = cars['Front_Track'].astype(float)
cars['Rear_Track'] = cars['Rear_Track'].astype(str)
cars['Rear_Track'] = [i.replace(' mm','') for i in cars['Rear_Track']]
cars['Rear_Track'] = [i.replace(' (R13)','') for i in cars['Rear_Track']]
cars['Rear_Track'] = [i.replace(',','') for i in cars['Rear_Track']]
cars['Rear_Track'] = cars['Rear_Track'].astype(float)
```

```python
cars['Wheelbase'] = cars['Wheelbase'].astype(str)
cars['Wheelbase'] = [i.replace(' mm','') for i in cars['Wheelbase']]
cars['Wheelbase'] = cars['Wheelbase'].astype(float)
```

```python
cars['City_Mileage'] = cars['City_Mileage'].fillna(cars['City_Mileage'].mean())
cars['Highway_Mileage'] = cars['Highway_Mileage'].fillna(cars['Highway_Mileage'].mean())
cars['ARAI_Certified_Mileage'] = cars['ARAI_Certified_Mileage'].fillna(cars['ARAI_Certified_Mileage'].mean())
cars['Kerb_Weight'] = cars['Kerb_Weight'].fillna(cars['Kerb_Weight'].mode())
cars['Length'] = cars['Length'].fillna(cars['Length'].mean())
cars['Height'] = cars['Height'].fillna(cars['Height'].mean())
cars['Width'] = cars['Width'].fillna(cars['Width'].mean())
cars['Displacement'] = cars['Displacement'].fillna(cars['Displacement'].mean())
cars['Wheelbase'] = cars['Wheelbase'].fillna(cars['Wheelbase'].mean())
cars['Fuel_Tank_Capacity'] = cars['Fuel_Tank_Capacity'].fillna(cars['Fuel_Tank_Capacity'].mean())
cars.isnull().sum()
```

1.From the price boxplot it is clear that The brands with the most expensive vehicles in the dataset belong to Bmw,Buick,Jaguar and porsche.
2. Whereas the lower priced cars belong to chevrolet
3. The median price of gas vehicles is lower than that of Diesel Vehicles.
4. 75th percentile of standard aspirated vehicles have a price lower than the median price of turbo aspirated vehicles. 
5. Two and four Door vehicles are almost equally priced. There are however some outliers in the price of 
two-door vehicles. 
6. Hatchback vehicles have the lowest median price of vehicles in the data set whereas hardtop vehicles have 
the highest median price.
7. The price of vehicles with rear placed engines is significantly higher than the price of vehicles with front 
placed engines. 
8. Almost all vehicles in the dataset have engines placed in the front of the vehicle. However, the price of 
vehicles with rear placed engines is significantly higher than the price of vehicles with front placed engines. 
9. The median cost of eight cylinder vehicles is higher than other cylinder categories.
10. It is clear that vehicles Multi-port Fuel Injection [MPFI] fuelsystem have the highest median price. There are 
also some outliers on the higher price side having MPFI systems.
11. Vehicles with OHCV engine type falls under higher price range.
